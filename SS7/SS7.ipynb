Counting Sort

Bài toán: Counting Sort giải quyết bài toán sắp xếp dãy có ít giá trị riêng biệt

```python
import matplotlib.pyplot as plt

def show_list(arr, title=''):
    fig, ax = plt.subplots(figsize=(20, 4)) # Tạo một figure (khung chính) và axes (trục biểu đồ).
    ax.bar(range(len(arr)), arr) # ax là đối tượng Axes, đại diện cho khu vực biểu đồ 
    ax.set_title(title) 
    ax.set_xlabel('index')
    ax.set_ylabel('value')
    ax.set_xticks(range(len(arr)))
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.show()
```

## Counting Sort
Bài toán: Cho một dãy gồm 0 < n < 106 số nguyên, mỗi số có giá trị trong khoảng [0, 99], sắp xếp dãy trên theo thứ tự tăng dần.

```python
number_list = [5, 9, 1, 12, 0, 10, 15, 0, 0, 20, 75]
show_list(number_list, title='number_list')
```

Cách tiếp cận: 
1. Duyệt qua dãy để đếm số lần xuất hiện của từng giá trị.
2. Sử dụng các giá trị đếm được để đưa các phần tử vào dãy mới theo thứ tự đã sắp xếp.

VD: Ta cần sx dãy [1,4,1,2,7,5,2] chứa các giá trị từ 0 ->7
Bước 1. Tạo mảng count chứa số lần xuất hiện của mỗi phần tử
       i: 0 1 2 3 4 5 6 7
count[i]: 0 2 2 0 1 1 0 1
=> 1 1 2 2 4 5 7

Bước 2: Duyệt count theo thứ tự từ đầu đến cuối để đưa các giá trị vào dãy mới.

```python


def counting_sort(arr):
    if len(arr) <= 1:
        return 
    
    max_val = max(arr) ## lấy ra giá trị lớn nhất: 7
    
    count = [0] * (max_val+1) # danh sách gồm 8 số 0
    ## Xử lý count
    for el in arr:
        count[el] += 1
     
    index = 0
    for i in range(max_val+1): #i: 0 - 7
        for j in range(count[i]): #j: 0 - 1
            arr[index] = i
            index += 1
            
            
```

```python
number_list_2 = number_list[:]
counting_sort(number_list_2)

print("Before sorting: {}".format(number_list))
print("After sorting: {}".format(number_list_2))
show_list(number_list_2, "number_list_2")
```
